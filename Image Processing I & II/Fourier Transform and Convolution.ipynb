```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
newton = np.asarray(Image.open('newton.jpeg'))
newton = newton/255

forest = np.asarray(Image.open('wallpaper.jpeg'))
forest = forest/255
```

```python
def inverseFourierImgShow(freqR, freqG, freqB):
    ir = np.fft.ifft2(freqR)
    ig = np.fft.ifft2(freqG)
    ib = np.fft.ifft2(freqB)
    
    ar = (ir.real**2 + ir.imag**2)**0.5
    ag = (ig.real**2 + ig.imag**2)**0.5
    ab = (ib.real**2 + ib.imag**2)**0.5
    
    colorImg = np.zeros((ar.shape[0], ar.shape[1], 3))
    colorImg[:, :, 0] = ar
    colorImg[:, :, 1] = ag
    colorImg[:, :, 2] = ab
    
    figure(figsize=(8, 6), dpi=80)
    plt.imshow(colorImg)
    plt.show()
    return colorImg

def getFourierFreqFun(img):
    return np.fft.fft2(img[:, :, 0]), np.fft.fft2(img[:, :, 1]), np.fft.fft2(img[:, :, 2])

def showFreqFun(freq):
    img = np.log10(((freq.real**2 + freq.imag**2) **0.5) + 0.001)
    img /= np.max(img)
    
    figure(figsize=(8, 6), dpi=80)
    plt.imshow(img)
    plt.show()
```

```python
nfr, nfg, nfb = getFourierFreqFun(newton)
ffr, ffg, ffb = getFourierFreqFun(forest)
r = 300

#replace frequency coefficients
nfr[0:r, 0:r] = ffr[0:r, 0:r]
nfg[0:r, 0:r] = ffg[0:r, 0:r]
nfb[0:r, 0:r] = ffb[0:r, 0:r]

inverseFourierImgShow(nfr, nfg, nfb)

```
