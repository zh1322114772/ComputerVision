```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import math
```

```python
#load image
img = np.asarray(Image.open('newton.jpeg'))/255
img
```

```python
figure(figsize=(10, 8), dpi=80)
plt.imshow(img)
plt.show()
```

```python
def gaussianBlur(img, blurRadius):
    
    #generate 1d gaussian function with sigma of 1
    gaussianKernel = np.zeros((blurRadius * 2) + 1)
    #simpling interval
    delta_x = 6/((blurRadius * 2) + 1)
    
    for x in range(0, (blurRadius * 2) + 1):
        x_pos = -3 + (x * delta_x) + (delta_x/2)
        #get y value of gaussian function
        gaussianKernel[x] = (1/(((2*math.pi) ** 0.5))) * (math.e ** (-((x_pos ** 2)/2))) * delta_x
    
    #padding edge to avoid out of index
    ImgA = np.zeros((img.shape[0] + (blurRadius * 2), img.shape[1] + (blurRadius * 2), 3))
    ImgB = np.zeros((img.shape[0] + (blurRadius * 2), img.shape[1] + (blurRadius * 2), 3))
    ImgC = np.zeros((img.shape[0] + (blurRadius * 2), img.shape[1] + (blurRadius * 2), 3))
    
    #copy img to ImgA
    ImgA[blurRadius : img.shape[0] + blurRadius, blurRadius : img.shape[1] + blurRadius, :] = img[:, :, :]
    
    #horizontal convolution
    for y in range(blurRadius, img.shape[0] + blurRadius):
        for x in range(0, img.shape[1]):
            ImgB[y, blurRadius + x, 0] = np.dot(ImgA[y, x : (x + gaussianKernel.shape[0]), 0].T, gaussianKernel)
            ImgB[y, blurRadius + x, 1] = np.dot(ImgA[y, x : (x + gaussianKernel.shape[0]), 1].T, gaussianKernel)
            ImgB[y, blurRadius + x, 2] = np.dot(ImgA[y, x : (x + gaussianKernel.shape[0]), 2].T, gaussianKernel)
    
    #vertical convolution
    for x in range(blurRadius, img.shape[1] + blurRadius):
        for y in range(0, img.shape[0]):
            ImgC[blurRadius + y, x, 0] = np.dot(ImgB[y : (y + gaussianKernel.shape[0]), x, 0], gaussianKernel)
            ImgC[blurRadius + y, x, 1] = np.dot(ImgB[y : (y + gaussianKernel.shape[0]), x, 1], gaussianKernel)
            ImgC[blurRadius + y, x, 2] = np.dot(ImgB[y : (y + gaussianKernel.shape[0]), x, 2], gaussianKernel)
    
    #remove edge and return
    return ImgC[blurRadius : img.shape[0] + blurRadius, blurRadius : img.shape[1] + blurRadius, :]
```

```python
%%time
#blur radius 5
figure(figsize=(10, 8), dpi=80)
img1 = gaussianBlur(img, 5)
plt.imshow(img1)
plt.show()
```
